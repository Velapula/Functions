# 1. What is the difference between a function and a method in Python?

-> Function- A function is a block of reusable code, defined using the def keyword, that performs a specific task. It is not associated with any object and can be called independently.

Method- A method is a function that is associated with an object -usually part of a class. Methods are functions that belong to a class and typically operate on the object's data. They are called using an instance of the class.

```python
# Example for Function:
def greet(name):
    return f"Hello, {name}!"

print(greet("Sunil"))  #calling the function

# Example for Method:
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Sunil")
print(p.greet())  # Calling the method on an object
```

# 2. Explain the concept of function arguments and parameters in Python.
-> In Python, parameters and arguments play a crucial role in defining and calling functions.

Parameters are the variables listed inside the function definition, acting as placeholders for values that the function will receive when called.

On the other hand, arguments are the actual values passed to the function when it is invoked, replacing the parameters.

Python allows several types of arguments:

positional arguments- where values are assigned based on their order in the function call.

keyword arguments- where values are assigned using parameter names, making the order irrelevant.

default arguments- where a parameter is given a default value if no argument is provided.

variable-length arguments- which allow a function to accept an arbitrary number of arguments.

The *args syntax is used to accept multiple positional arguments as a tuple, while **kwargs is used for multiple keyword arguments, storing them as a dictionary. These flexible argument-passing mechanisms make Python functions highly versatile and dynamic.


```python
def introduce(name, age=25, *hobbies, **details): #name -Positional argument (Required),hobbies-Variable length positional arguments (Stores multiple values as a tuple)


    print(f"My name is {name} and I am {age} years old.") #age=25 - Default argument uses 25 if not provided


    if hobbies:
        print(f"My hobbies are: {', '.join(hobbies)}.")
    if details:
              print(f"My hobbies are: {', '.join(hobbies)}.")

    if details:
        print("Additional details:", details)#details â†’ Variable-length keyword arguments (Stores additional named values in a dictionary)

# Calling the function with different types of arguments
introduce("Sunil")  # Default age is used
introduce("sunny", 30, "Reading", "Games")  # Positional and *args
introduce("Raj", city="Hyderabad", profession="Engineer")  # **kwargs
introduce("Deva", 22, "Traveling", "Photography", country="Canada", language="English")
```

# 3. What are the different ways to define and call a function in Python?


-> In Python, functions can be defined and called in multiple ways based on their use case. Below are the various types of function definitions and how they can be called.

```python
# Normal Function
def greet(name):
    return f"Hello, {name}!"
print(greet("Sunil"))

#Default Arguments
def greet(name="Guest"):
    return f"Hello, {name}!"
print(greet())  # Uses default value

# Variable-Length Arguments (*args, **kwargs)
def sum_numbers(*args):
    return sum(args)
print(sum_numbers(1, 2, 3, 4))  # Output: 10

# Lambda (Anonymous) Function
add = lambda x, y: x + y
print(add(5, 5))

# Nested Function
def outer():
    def inner():
        return "Hello from inner!"
    return inner()
print(outer())

# Higher-Order Function
def apply_operation(func, x, y):
    return func(x, y)
print(apply_operation(lambda a, b: a * b, 6, 6))

# Recursion
def factorial(n):
    return 1 if n == 1 else n * factorial(n - 1)
print(factorial(7))

# Class Methods
class Person:
    def __init__(self, name):
        self.name = name
    def greet(self):
        return f"Hello, my name is {self.name}."
p = Person("Sunil")
print(p.greet())

"""""Python allows number of ways to define and call functions,
providing flexibility based on the programming needs.
Whether it's a normal function, lambda function, recursive function, or higher-order function,
Python's function-handling capabilities make it powerful and efficient."""
```

# 4. What is the purpose of the "return" statement in a Python function?

-> The return statement in Python is used to exit a function and send a value back to the caller. It allows functions to produce results that can be stored, used, or manipulated later.

If a function does not use return, it automatically returns None. This means the function performs its task but does not send back a result to the caller.
Below are the examples:

```python
# Returning a Value
def sub(a, b):
    return a - b

result = sub(3, 3)
print(result)

# Returning Without a Value
def greet(name):
    print(f"Hello, {name}!")
    return  # Explicitly returns None

print(greet("Sunil"))

# Using return to Exit a Function Early
def check_even(n):
    if n % 2 == 0:
        return "Even"
    return "Odd"

print(check_even(4))

# Function Without return
def greet(name):
    print(f"Hello, {name}!")

result = greet("Sunil")  # No return statement
print(result)  # Output: Hello, Alice! \n None

#Conclusion:
#return sends a result from a function.
#It stops execution of the function.
#If omitted, the function returns None.
#Can return multiple values as a tuple.
#Functions without return are useful for tasks like printing, modifying data, or logging, but not for calculations.
```

# 5. What are iterators in Python and how do they differ from iterables?

->In Python, iterables and iterators are key concepts used for looping over collections of data, but they differ in functionality.

An iterable is any object, like a list, tuple, or string, that can be traversed using a loop. It implements the __iter__() method, which returns an iterator.

An iterator, on the other hand, is an object that produces values one at a time using the __next__() method, allowing efficient memory usage by generating data on demand instead of storing it all at once.

While an iterable can be looped over multiple times, an iterator maintains its internal state and is exhausted once all elements are retrieved.

Python provides the iter() function to convert an iterable into an iterator, allowing manual iteration using next(). Additionally, custom iterators can be created by defining a class with __iter__() and __next__().

Iterators are particularly useful when working with large datasets, as they support lazy evaluation, fetching data only when needed, which optimizes memory usage and performance.

Key Differences Between Iterables and Iterators-

Iterable implements __iter__() while Iterators implements __iter__() and __next__()

Iterable stores all elements in memory and Iterators retrieves elements one at a time (saves memory).

Examples of Iterable are Lists, Tuples, Sets, Strings	 and for Iterator-Objects from iter()


```python
# Examples of Iterables:
my_list = [1, 2, 3, 4]
for item in my_list:
    print(item) # Here, my_list is an iterable because it can be looped again.
```

```python
# Example of an Iterator:
my_iter = iter(my_list)  # Convert list into an iterator
print(next(my_iter))
print(next(my_iter))
```

# 6. Explain the concept of generators in Python and how they are defined.

-> In Python, generators are a special type of iterator that allows for efficient iteration by generating values on the fly instead of storing them in memory.

They are defined using functions with the yield keyword, which pauses the function's execution and remembers its state, allowing it to resume from where it left off when called again.

Unlike normal functions that return a single value and terminate, generators produce a sequence of values lazily, making them ideal for handling large datasets efficiently.

A generator is defined like a regular function but uses yield instead of return. Each time yield is encountered, the function pauses and sends the value back to the caller, resuming execution on the next call.

Since generators do not store values in memory but generate them on demand, they are memory-efficient and useful for tasks like reading large files, processing real-time data streams, or implementing infinite sequences.

```python
# For example:
def count_up(n):
    for i in range(1, n + 1):
        yield i  # Pauses execution and returns the next value

counter = count_up(5)
print(next(counter))
print(next(counter))
```

# 7. What are the advantages of using generators over regular functions?

-> Generators offer several advantages over regular functions, especially when dealing with large datasets or continuous data streams.

One major benefit is memory efficiency, as generators yield values one at a time instead of storing them all in memory, making them ideal for handling large files or infinite sequences.

They also improve performance by enabling lazy evaluation, meaning values are computed only when needed, reducing unnecessary computations.

Generators maintain state between function calls using the yield keyword, allowing them to resume execution from where they left off without restarting.

This makes them useful for tasks like streaming data or implementing iterators without writing complex state-tracking logic.

Additionally, generators offer cleaner, more readable code compared to manually implementing iterators with __iter__() and __next__().

They also work well with pipelines, where data is processed step by step without storing intermediate results, further improving efficiency.

Overall, generators enhance performance, save memory, and simplify code structure, making them a powerful tool in Python programming.

```python
# Regular Function (Consumes More Memory)
def square_numbers(n):
    result = []
    for i in range(n):
        result.append(i * i)  # Stores all values in a list
    return result

print(square_numbers(5))


# Generator Function (Memory Efficient)
def square_numbers_gen(n):
    for i in range(n):
        yield i * i  # Generates values one by one

gen = square_numbers_gen(5)  # Creating a generator
print(next(gen))
print(next(gen))
print(next(gen))
```

#9. Explain the purpose and usage of the `map()` function in Python.

-> The map() function in Python is used to apply a function to each item in an iterable (e.g., list, tuple, etc.) and return a new iterable (a map object).

This is useful for transforming data without using explicit loops, making the code cleaner and more efficient.

Syntax of map()- map(function, iterable)

```python
# Example 1: Using map() to Square Numbers-
numbers = [1, 2, 3]
squared = map(lambda x: x * x, numbers)
print(list(squared))

# Example 2: Using map() with a Regular Function-
def double(x):
    return x * 2
numbers = [1, 2, 3]
doubled = map(double, numbers)
print(list(doubled))

# Example 3: Mapping Multiple Iterables-
a = [1, 2, 3]
b = [4, 5, 6]
sum_list = map(lambda x, y: x + y, a, b)
print(list(sum_list))
```

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> The functions map(), reduce(), and filter() in Python are built-in higher-order functions that allow for functional programming logic by applying functions to iterables.

1. map()

Purpose: Transforms each element in an iterable by applying a function to it.

Returns: A map object (an iterator) containing the transformed elements.

2. reduce()

Purpose: Reduces an iterable to a single value by applying a function cumulatively.

Returns: A single result after applying the function repeatedly.

Requires: from functools import reduce

3. filter()

Purpose: Filters elements in an iterable based on a condition.

Returns: A filter object (an iterator) containing only elements that satisfy the condition.

```python
# Examples

#map()
numbers = [1, 2, 3]
squared = map(lambda x: x**2, numbers)
print(list(squared))

# reduce()
from functools import reduce
numbers = [1, 2, 3]
sum_all = reduce(lambda x, y: x + y, numbers)
print(sum_all)

#filter()
numbers = [1, 2, 3, 4, 5]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):  # Check divisibility from 2 to n-1
        if n % i == 0:
            return False
    return True
primes = [num for num in range(1, 201) if is_prime(num)]  # Get prime numbers from 1 to 200
print(primes)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, n):
        self.n, self.a, self.b = n, 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.n <= 0:
            raise StopIteration
        self.n -= 1
        self.a, self.b = self.b, self.a + self.b
        return self.a
for num in Fibonacci(10):
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exp):
    for i in range(exp + 1):  # Loop from 0 to exp
        yield 2 ** i  # Yield 2^i
for num in powers_of_two(5):
    print(num, end=" ")
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")

# Example usage
file_path = "example.txt"
for line in read_file_line_by_line(file_path):
    print(line)
```
